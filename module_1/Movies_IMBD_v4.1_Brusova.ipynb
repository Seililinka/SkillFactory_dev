```python
## второй вариант решения
import itertools
from itertools import combinations
import pandas as pd
data_49=data_copy[['ds1','ds2','ds3','ds4','ds5']] = data.cast.str.split("|",expand=True)
## создаем датасет с разбиением актеров горизонтально по столбцам 
new=['cast1','cast2','cast3','cast4','cast5'] 
data_49.set_axis(new,axis='columns',inplace=True)
## очищаем полученный датасет от нулевых значений
data_49['cast1']=data_49['cast1'].fillna(' ')
data_49['cast2']=data_49['cast2'].fillna(' ')
data_49['cast3']=data_49['cast3'].fillna(' ')
data_49['cast4']=data_49['cast4'].fillna(' ')
data_49['cast5']=data_49['cast5'].fillna(' ')
df=data_49
df['comb'] = df.apply(lambda r: list(combinations(r, 2)),axis=1) ## создаем колонку с парами актеров
df_1 = df.reset_index()
df_1['comb1']=df_1.comb.str.split(')') ## разбиваем пары по разделителю ')'
df_1=df_1.explode('comb') ## формируем столбец разнося пары по вертикали в один столбец 
df_1['count']=1 ## добавляем колонку счетчика
df_2=df_1.groupby(['comb'])['count'].sum() ## группируем одинаковые пары
df_2 = df_2.reset_index()
df_2.columns = ['ds','count']
display(df_2.sort_values('count',ascending=False)) ## сортируем по убыванию количества в паре
```
