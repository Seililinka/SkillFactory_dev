```python
import pandas as pd
import glob
import os
```

```python
# path_all='/home/msi/Документы/project/seat_belt/train/RESULTS_OUT/seat_belt_2021-09-03_16-12-42_mobilenet_v3_large/'
path_all='/home/msi/Документы/project/car_classification/RESULT_OUT/cars_classific_2021-09-12_18-34-03_resnet18/'

experiments=glob.glob(f'{path_all}/*')

df_dict={}
df_concat = pd.DataFrame()

for i in experiments:
    df_tmp=pd.read_csv(os.path.join(i,'result.csv'),sep='\t')
    del df_tmp['Unnamed: 0']
    
    
    
    df_param=pd.read_csv(os.path.join(i,'param_expirement.csv'),sep='\t')
    del df_param['Unnamed: 0']
    df_param['hash_exprmnt']=os.path.basename(i)[:8]
    df_param=pd.concat([df_param]*len(df_tmp), ignore_index=True)
    
    df_tmp=pd.concat([df_tmp, df_param], axis=1)
    
    df_dict[os.path.basename(i)[:8]]=df_tmp.copy()
    df_concat = df_concat.append(df_tmp.copy(), ignore_index=True)
    del df_tmp
```

```python
df_concat.sort_values(by='valid_loss')
```
